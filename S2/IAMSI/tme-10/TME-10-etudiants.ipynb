<font size="+1">**[Q]**</font> Ecrire la fonction *chargeBase* qui prend en argument un nom de fichier, respectant le format énoncé plus haut, le lit et rend un dictionnaire dont les clés sont les numéros de ligne (une transaction) et la valeur associée est l'itemset correspondant représenté sous forme d'un ensemble Python.

Ici, il est plus intéressant de représenter un itemset comme un ensemble d'items (en Python, on utilise les *set()*) plutôt que comme une liste. Il sera ainsi plus facile de réaliser des comparaisons d'ensembles ou des ajouts d'éléments.

Dans le reste de ce document, on appelle **BASE** un dictionnaire de ce type.

```python
def chargeBase(fileName):
    with open(fileName, 'r') as fichier:
        lecteur = csv.reader(fichier, delimiter=' ')
        i = 0
        dict = {}
        for ligne in lecteur:
            i += 1
            dict[i] = set(ligne)

        return dict
```

```python
print("Résultat du chargement de 'exemple-1.txt', on obtient: ")
Base1 = chargeBase('./datasets/exemple-1.txt')
Base1
```

On utilise la variable **Base1** dans la suite pour faire référence à cette base.

## Itemsets et support

<font size="+1">**[Q]**</font> Ecrire la fonction *noms_items* qui prend en argument une BASE et rend l'ensemble des items qui composent cette base.

```python
def noms_items(base):
    res = set()
    for value in base.itervalues():
        res = res | value
    
    return res
```

```python
print("Pour la BASE précédente:")
noms_items(Base1)
```

<font size="+1">**[Q]**</font> Ecrire la fonction *singletons* qui prend en argument une BASE et rend l'ensemble des itemsets de taille 1 obtenus à partir de cette base.

```python
def singletons(base):
    res = []
    items = noms_items(base)
    for i in items:
        s = set()
        s.add(i)
        res.append(s)
    return res
```

```python
print("Exemple: singletons(Base1) rend:")
singletons(Base1)
```

<font size="+1">**[Q]**</font> Ecrire la fonction *comptage* qui, pour une BASE donnée et un itemset donné, rend le nombre de fois où cet itemset apparaît dans la BASE.

```python
def comptage(base, itemset):
    res = 0
    for value in base.itervalues():
        if itemset.issubset(value):
            res += 1
            
    return res
```

```python
print("Par exemple, comptage de l'itemset {'a','b','c'} dans la base précédente: ")
print("comptage(Base1,{'a','b','c'}) rend la valeur "+str(comptage(Base1,{'a','b','c'})))
```

<font size="+1">**[Q]**</font> Ecrire la fonction *support* qui, pour une BASE donnée et un itemset donné, rend le support de cet itemset dans la BASE.

```python
def support(base, itemset):
    cmp = comptage(base,itemset)
    return cmp/float(len(base))
```

```python
print("Support de l'itemset {'a','b','c'} dans la base précédente: "+str(support(Base1,{'a','b','c'})))
```

Vous allez réaliser une implémentation de la partie de construction des itemsets fréquents de l'algorithme a-priori. On ne s'intéressera pas (mais cela pourra être fait en complément) à la génération des règles d'association.

<font size="+1">**[Q]**</font> Ecrire la fonction *apriori_gen* qui prend en argument une liste d'itemsets de même longueur $k$, applique l'algorithme apriori-gen pour rendre la liste des itemsets candidats de longueurs $k+1$.

Remarque: attention, ici on utilise une **liste** pour stocker les itemsets (un itemset est un ensemble Python) car il n'est pas possible en Python de créer des ensembles d'ensembles.


```python
def apriori_gen(list_itemsets):
    res = []
    for itemset1 in list_itemsets:
        for itemset2 in list_itemsets:
            c = itemset1 | itemset2
            if len(c) == len(itemset1)+1:
                notGood = False
                for i in c:
                    temp = set(c)
                    temp.discard(i)
                    if temp not in list_itemsets:
                        notGood = True
                if not notGood:
                    if c not in res:
                        res.append(c)
    return res
                
```

```python
print("Exemple: apriori_gen([{'a'}, {'b'}, {'c'}, {'d'}]) rend ")
print(apriori_gen([{'a'}, {'b'}, {'c'}, {'d'}]))

print("\nExemple: apriori_gen([{'a', 'b'}, {'a', 'd'}, {'b', 'd'}, {'b', 'c'}, {'c', 'd'}]) rend ")
print(apriori_gen([{'a', 'b'}, {'a', 'd'}, {'b', 'd'}, {'b', 'c'}, {'c', 'd'}]))
```

<font size="+1">**[Q]**</font> Ecrire la fonction *apriori* qui prend en argument une BASE et une valeur réelle comprise entre 0 et 1, et qui rend une liste de tuples composés comme suit:
- le premier élément du tuple est un itemset fréquent trouvé
- le second élément du tuple est la valeur de support correspondant

```python
def apriori(base, minsup):
    res = []
    c = singletons(base)
    k = 1
    while c:
        f = []
        for i in c:
            sup = support(base,i)
            if sup>=minsup:
                f.append(i)
                res.append((i,sup))
        
        k += 1
        c = apriori_gen(f)
    
    return res
```

## Annexes: utilisation des programmes *apriori32* et *fpgrowth32*

La documentation complète de ces programmes est disponible aux URL suivantes:
- pour *apriori*: http://www.borgelt.net/doc/apriori/apriori.html
- pour *fp-growth*: http://www.borgelt.net/doc/fpgrowth/fpgrowth.html


Leur format d'utilisation général est:

        ./<programme> [options] infile [outfile]

(où *<programme>* est soit *apriori32*, soit *fpgrowth32*).

Les options de base de ces 2 programmes sont:
- sans argument: génération des itemsets fréquents (argument {\tt -ts} activé par défaut)
- *tr*: pour obtenir les règles d'associations
- *s*: pour fournir une valeur minimale de support
- *m*: pour fournir un nombre minimum d'items dans un itemset
- *n*: pour fournir un nombre maximum d'items dans un itemset


Par exemple (commandes lancées dans le répertoire père du répertoire *datasets/*):

        ./apriori32 -trs50m2n5 datasets/exemple-1.txt fichier-resultat.out  
        ./fpgrowth32 -trs50m2n5 datasets/mushrooms.txt fichier-resultat.out
